# Voting Methods

Let's prepare data

```python
from sklearn.datasets import make_moons
```

```python

X, y = make_moons(n_samples= 600, noise=0.2, random_state=12)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=226)
```

Let's build a voting classifier:

```python
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier

voting_clf = VotingClassifier(estimators=[
    ('sgd', SGDClassifier(random_state=622)),
    ('svc', SVC(random_state=622)),
    ('rf', RandomForestClassifier(random_state=622))
])

voting_clf.fit(X_train, y_train)
```

```python
for name, clf in voting_clf.named_estimators_.items():
    print(name, '=', clf.score(X_test, y_test))
```

```python
voting_clf.score(X_test, y_test)
```
